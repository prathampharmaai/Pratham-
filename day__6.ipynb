```python
import pandas as pd
import numpy as np
Url = 'https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv'
df = pd.read_csv(Url)
df.loc[10,'Molwt'] = np.nan
df.loc[50,'Molwt'] = np.nan
df.loc[100,'logS'] = np.nan

df.isnull().sum()
df['Molwt']=df['MolWt'].fillna(df['MolWt'].mean())
df.dropna(subset=['logS'])
```
