```python
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')

All_drugs_x = ['Aspirin','paracetamol','Ibupofen','Atenol','Metformin']
mw_y = ['180.2','206.3','151.2','266.3','129.2']

plt.plot(All_drugs_x,mw_y,color='red')


solublity = ['121.6','202.4','152','263','122']
plt.plot(All_drugs_x ,solublity,color='blue')

plt.xlabel('Drugs')
plt.ylabel('Mw')
plt.title('Mw of Drugs')
plt.show()
```

```python
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')

All_drugs_x = ['Aspirin','paracetamol','Ibupofen','Atenol','Metformin']
mw_y = ['180.2','206.3','151.2','266.3','129.2']
colors_list = ['blue','orange','green','red','purple']
plt.figure(figsize=(8,5))

plt.bar(All_drugs_x,mw_y,color=colors_list)


plt.xlabel('Drugs')
plt.ylabel('Mw')
plt.title('Mw of Drugs')
plt.show()
```
